```python
#AUTO WHITE BALANCE(SIMPLE)-
wb = cv2.xphoto.createSimpleWB()
wb.setP(0.3)#The percentage of top and bottom values to be ignored
```

```python
#Edge Enhancement and Auto white balance-
final_image=[]
i='WithMask'
W1 = os.path.join(P1,i) #Joins the directory and specific file path
print(W1)
count=0
for img in os.listdir(W1): # Find image in path 
    image = cv2.resize(cv2.imread(os.path.join(W1,img)), (64,64))#resize image
    image1 = wb.balanceWhite(image)# White balance on raw image
    blurred = cv2.GaussianBlur(image1, (9,9), 8)# Blurring the image with standard deviation of 8 along x and y 
    imagefin = cv2.addWeighted(image1, 1.0 + 4.0, blurred, -4.0, 0) #
    final_image.append (np.ndarray.flatten(imagefin)) #Convert an image matrix to a vector  
    count=count+1
    if(count==300):
        break
df = pd.DataFrame(final_image)
df = df.assign(Mask = '1') #Assigning the label 1 for with mask
df
```

```python
final_image_without=[]
i='WithoutMask'
W1 = os.path.join(P1,i) #Joins the directory and specific file path
print(W1)
count2=0
for img in os.listdir(W1): # Find image in path
    image = cv2.resize(cv2.imread(os.path.join(W1,img)), (64,64))
    image1 = wb.balanceWhite(image)
    blurred = cv2.GaussianBlur(image1, (9,9), 8)
    imagefin = cv2.addWeighted(image1, 1.0 + 4.0, blurred, -4.0, 0)
    final_image_without.append (np.ndarray.flatten(imagefin)) #Convert an image matrix to a vector
    count2 = count2+1
    if(count2==300):
        break
df_without = pd.DataFrame(final_image_without)
df_without= df_without.assign(Mask= '0') #Assigning the label 0 for without mask
df_without
```

```python
import keras
```

```python
from keras.utils import np_utils
CatoMask = keras.utils.to_categorical(dfcls, num_classes=2) #One hot encoding for categorical labels
```

```python
dfn=dfn.drop("Mask",1) #Removing the class variable
dfn
```

```python
imgarr.shape
```

```python
imgarr1=np.reshape(imgarr,(imgarr.shape[0],64,64,3)) # Reshaping the image according to dimensions needed for neural network
imgarr1.shape
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
Xtr,Xtt,Ytr,Ytt=train_test_split(imgarr1,CatoMask,test_size=0.1,random_state=0, stratify=CatoMask) # Splitting the data in a balanced fashion
```

```python
Ytr.shape
```

```python
import tensorflow as cnn
```

```python
model=cnn.keras.models.Sequential([
     cnn.keras.layers.Conv2D(32, (3,3), padding='same',activation='relu', input_shape=(64,64,3)),
     cnn.keras.layers.MaxPooling2D(2, 2),
     cnn.keras.layers.Dropout(0.1),
     cnn.keras.layers.Conv2D(64, (3,3), padding='same', activation='relu'),
     cnn.keras.layers.MaxPooling2D(2, 2),
     cnn.keras.layers.Dropout(0.1),
     cnn.keras.layers.Flatten(),
     cnn.keras.layers.Dense(512, activation='relu'),
     cnn.keras.layers.Dense(2, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(Xtr,Ytr, epochs = 9,validation_data = (Xtt,Ytt))
```

```python
model.evaluate(Xtt,Ytt)
```

```python
img_cap=cv2.VideoCapture(0)# Capture the face from the camera
style = cv2.FONT_HERSHEY_COMPLEX
while True:
    pres, face = img_cap.read()#Read the complete image captured at a time
    if pres:
        detect = HCCimg.detectMultiScale(face)# From the image detect the face portion 
        for x,y,w,h in detect:
            cv2.rectangle(face,(x,y),(x+w,y+h),(134,167,245),5) #Draw a rectangle on the face part
            face1 = face[ y:y+h , x:x+w , :]#Slice only the face portion of the image
            #Pre-processing similar to training data
            face2 = cv2.resize(face1,(96,96))
            face3 = np.ndarray.flatten(face2)
            face3=face3/255
            face4 = np.reshape(face3,(1,96,96,3))
            pred = model31.predict(face4) #Predicting the target variable
            if(pred[:,1] == 1 or pred[:,1]<1 and pred[:,1]>0.98):
                n='Mask'
                cv2.putText(face,n,(x,y),style,1,(234,233,167),2)#Displaying the corresponding test on screen
            else:
                n1='No Mask'
                cv2.putText(face,n1,(x,y),style,1,(234,233,167),2)#Displaying the corresponding test on screen
        cv2.imshow('Takevideo',face)#Display window to show images captured by camera
        if cv2.waitKey(2) == 27:
            break #The display window remains until escape key is pressed
img_cap.release()
cv2.destroyAllWindows() #Closing the display window
```
